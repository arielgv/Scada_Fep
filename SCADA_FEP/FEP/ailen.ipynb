```python
import requests
import pandas as pd
import time

def get_places(api_key, query, location, radius=100000):
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
    results = []

    params = {
        "query": query,
        "location": location,
        "radius": radius,
        "key": api_key
    }

    while True:
        response = requests.get(url, params=params)
        data = response.json()

        # Extraer lugares de los resultados
        for place in data.get('results', []):
            results.append({
                "Nombre": place.get('name'),
                "Dirección": place.get('formatted_address', 'No disponible')
            })

        # Verificar si hay un token de paginación para la siguiente página
        page_token = data.get('next_page_token')
        if page_token:
            # Google requiere una pequeña pausa antes de solicitar la siguiente página
            time.sleep(4)
            params['pagetoken'] = page_token
            # Asegúrate de remover el parámetro 'query' ya que no se usa con 'pagetoken'
            params.pop('query', None)
            params.pop('location', None)
            params.pop('radius', None)
        else:
            break

    return results

# Configura tus parámetros aquí
api_key = "AIzaSyBnHYX_I3Uio_JIAIL4l_N4nVn1TIlS5Og"
query = "Casa de suprimentos eletronicos no Brasil"
location = "-22.90642,-43.18223"  # Coordenadas de São Paulo, Brasil

# Obtener los lugares utilizando la función definida
places = get_places(api_key, query, location)

# Convertir los resultados en un DataFrame de pandas y guardarlos en un archivo CSV
df = pd.DataFrame(places)
df.to_csv("Electronicariojan.csv", index=False)

print(f"Guardado: {len(places)} lugares en 'lugares_rio de janeiro.csv'")
```
